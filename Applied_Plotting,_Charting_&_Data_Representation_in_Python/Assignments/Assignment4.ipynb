I scraped the data values for daily Temperature (T_max, T_min, T_med) for daily Precipitation (pp, pp acum/month) and daily Relative Humidity (Humid_med) for the years 2000 to 2019 with every single day as row and a column for each variable measured.

After the DataFrame is mounted I run the following plotting script:

```python
def comparative_plot(df, year0, year1, yearA, yearB):
    
    years_old = np.arange(year0,year1)
    years_new = np.arange(yearA,yearB)
    
    month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    month_dict = {'Jan': 0, 'Feb': 1, 'Mar': 2, 'Apr': 3,
               'May': 4, 'Jun': 5, 'Jul': 6, 'Aug': 7,
               'Sep': 8, 'Oct': 9, 'Nov': 10, 'Dec': 11}
    
    var_dict = {'temp': ['T_max', 'T_min', 'T_med'], 'pp': ['Prec Diaria', 'Prec Acum Mensual'], 'h': ['Humid_med']}
    
    measures = ['T_max', 'T_min', 'T_med', 'Prec Diaria', 'Prec Acum Mensual', 'Humid_med']
    cols = ['Year', 'Month', 'Day'] + measures
        
    
    plt.style.use('grayscale')
    fig, ((ax1,ax2,ax3)) = plt.subplots(3, 1, sharex=True, figsize=(30,14))
        
    x_sh = [[-10,40], [0,100], [0,100]]
    
    ts = df[cols].copy()
    
    ts_old = ts.loc[ts['Year'].isin(years_old)].copy()
    ts_new = ts.loc[ts['Year'].isin(years_new)].copy()
    
    ts_old[var_dict['temp']] = ts_old[var_dict['temp']].rolling(window=7).mean()
    ts_new[var_dict['temp']] = ts_new[var_dict['temp']].rolling(window=7).mean()
        
    ts_old[var_dict['h']] = ts_old[var_dict['h']].replace(0, np.nan).dropna()
    ts_new[var_dict['h']] = ts_new[var_dict['h']].replace(0, np.nan).dropna()
        
    
    ts_old['Nmonth'] = ts_old['Month'].map(month_dict)
    ts_new['Nmonth'] = ts_new['Month'].map(month_dict)
        
    ts_old = (ts_old.groupby(['Nmonth', 'Day'])[measures]
            .mean()
            .reset_index()
            .dropna()
            .sort_values(['Nmonth', 'Day']))
    ts_new = (ts_new.groupby(['Nmonth', 'Day'])[measures]
            .mean()
            .reset_index()
            .dropna()
            .sort_values(['Nmonth', 'Day']))

    t_graph_old = ts_old[var_dict['temp']]
    t_graph_new = ts_new[var_dict['temp']]
    
    t_graph_old['T_max'].plot(color='maroon', label='Max old', alpha=0.8, linewidth=7, ax=ax1)
    t_graph_old['T_min'].plot(color='midnightblue', label='Min old', alpha=0.8, linewidth=7, ax=ax1)
    t_graph_old['T_med'].plot(color='darkolivegreen', label='Med old', alpha=0.8, linewidth=7, ax=ax1)
            
    t_graph_new['T_max'].plot(color='red', label='Max recent', linewidth=5, ax=ax1)
    t_graph_new['T_min'].plot(color='blue', label='Min recent', linewidth=5, ax=ax1)
    t_graph_new['T_med'].plot(color='green', label='Med recent', linewidth=5, ax=ax1)
        
    ax1.fill_between(ts_new.index, ts_new['T_max'], ts_old['T_max'], facecolor='red', alpha=0.4)
    ax1.fill_between(ts_new.index, ts_new['T_min'], ts_old['T_min'], facecolor='blue', alpha=0.4)
    ax1.fill_between(ts_new.index, ts_new['T_med'], ts_old['T_med'], facecolor='green', alpha=0.4)
        
    ax1.axis([0,365,x_sh[0][0],x_sh[0][1]])
    ax1.set_ylabel('Temperature measurements [°C]')
    plt.grid() 
    
    
    ts_old['Prec Diaria'].plot(color='midnightblue', label='pp day old', kind='bar', ax=ax2) 
    ts_old['Prec Acum Mensual'].plot(color='darkolivegreen', label='pp sum/month old', kind='bar', alpha=0.65, ax=ax2)

    ts_new['Prec Diaria'].plot(color='blue', label='pp day recent', kind='bar', ax=ax2)
    ts_new['Prec Acum Mensual'].plot(color='green', label='pp sum/month recent', kind='bar', alpha=0.85, ax=ax2)

    ax2.axis([0,365,x_sh[1][0],x_sh[1][1]])
    ax2.set_ylabel('Precipitation [mm]')
    plt.grid()
    
    
    ts_old['Humid_med'] = ts_old['Humid_med'].replace(0, np.nan).dropna()
    ts_new['Humid_med'] = ts_new['Humid_med'].replace(0, np.nan).dropna()

    ts_old['Humid_med'].plot(color='firebrick', label='%HR old', style='.', markersize=9, ax=ax3)
    ts_new['Humid_med'].plot(color='blue', label='%HR recent', style='.', markersize=9, ax=ax3)

    ax3.axis([0,365,x_sh[2][0],x_sh[2][1]])
    ax3.set_ylabel('Relative Humidity [%]')
        
    ax3.set_xlabel('Time')

    ax1.legend()
    ax2.legend()
    ax3.legend()
    ax1.grid()
    ax2.grid()
    ax3.grid()
    
    name = {'temp': 'Temperature', 'pp': 'Precipitation', 'h': 'Relative Humidity'}
    
    ax1.set_title('Average values comparison for La Reina, Santiago, Chile between {}-{} and {}-{}'
                 .format(year0, year1, yearA, yearB))
            
    _ = plt.xticks([0,31,58,89,119,149,180,211,241,272,302,333], month_names)
    
        
        
    plt.rc('axes', titlesize=35)    
    plt.rc('axes', labelsize=20)    
    plt.rc('xtick', labelsize=20) 
    plt.rc('ytick', labelsize=20)
    plt.rc('legend', fontsize=20)    
    plt.rc('figure', titlesize=30)
    
    #plt.savefig('Assignment 4 - La Reina weather')
    
    plt.show()
    
comparative_plot(df, 2000, 2015, 2015, 2020)
```
